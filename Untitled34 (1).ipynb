```python
import cv2
import mediapipe as mp
import numpy as np
```

```python

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose # pose estimation is imported from mediapipe
mp_drawing = mp.solutions.drawing_utils
# components in mediapipe are refered as solutions
```

```python
#video feed
cap=cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame=cap.read()
    cv2.imshow('mediapipe feed',frame)

    if cv2.waitKey(10) & 0xff ==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python

cap = cv2.VideoCapture(0)
## Setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
      
        # Make detection
        results = pose.process(image)
    
        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        POSE_LANDMARKS = {
    0: "Nose", 1: "Left Eye", 2: "Right Eye", 3: "Left Ear", 4: "Right Ear",
    5: "Left Shoulder", 6: "Right Shoulder", 7: "Left Elbow", 8: "Right Elbow",
    9: "Left Wrist", 10: "Right Wrist", 11: "Left Hip", 12: "Right Hip",
    13: "Left Knee", 14: "Right Knee", 15: "Left Ankle", 16: "Right Ankle",
    17: "Left Heel", 18: "Right Heel", 19: "Left Foot Index", 20: "Right Foot Index",
    21: "Left Eye Inner", 22: "Right Eye Inner", 23: "Left Eye Outer", 24: "Right Eye Outer",
    25: "Left Brow Inner", 26: "Right Brow Inner", 27: "Left Brow Outer", 28: "Right Brow Outer",
    29: "Left Cheek", 30: "Right Cheek", 31: "Mouth Left", 32: "Mouth Right"
}

        try:
            landmarks=results.pose_landmarks.landmark
            print(landmarks)
        except:
            pass

        

        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
   
        
        cv2.imshow('Mediapipe Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

```

lets find out angle b/w left shoulder,then joint(elbow) and hand(wrist) [11,13,15]

```python
def calculate_angle(a,b,c):
    a=np.array(a)#1st 11 
    b=np.array(b)#mid 13
    c=np.array(c)#end 15

    radians=np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0])
    # radians = angle b/w vector(b->c) and vector(b->a) so common pt b and 2 lines b->c and b->a
    # of x,y,z we only use wt needed based on requirement co-ordinate indxnig 0:x,1:y,2:z
    
    angle=np.abs(radians*180/np.pi)
    #converts radians to degrees.
    

    if angle>180.0:
        angle=360-angle
    return angle
```
